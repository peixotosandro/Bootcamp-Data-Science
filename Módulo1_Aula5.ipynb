```python
dados_populacao_2019 = pd.read_csv("/content/drive/MyDrive/BootcampDataScience/A194744189_28_143_208.csv",encoding="ISO-8859-1", skiprows=3, 
                    sep=";",skipfooter=12, thousands=".", decimal=",")
```

```python
dados_populacao_2019 = dados_populacao_2019.set_index("Unidade da Federação")
dados_populacao_2019.sort_values('População_estimada',ascending=False)
```

```python
dez_2019 = ordenado_por_total.columns[142]
dez_2019
```

```python
invest_dez_2019 = ordenado_por_total[dez_2019]
invest_dez_2019 = invest_dez_2019 * 1000000
invest_dez_2019
```

```python
uf_invest = list(map(lambda uf: invest_dez_2019.loc[uf], dados_populacao_2019.index))
invest_dez_2019_pop = dados_populacao_2019.copy()
invest_dez_2019_pop['Total'] = uf_invest
invest_dez_2019_pop
```

```python
invest_dez_2019_pop['Invest/Pessoa'] = invest_dez_2019_pop['Total'] / invest_dez_2019_pop['População_estimada']
ordenado_por_invest_pessoa_dez_2019 = invest_dez_2019_pop.sort_values('Invest/Pessoa', ascending=False) 
ordenado_por_invest_pessoa_dez_2019
```

```python
ordenado_por_invest_pessoa_dez_2019['Invest/Pessoa'].T.plot(kind='bar',figsize=(15,6)).grid()
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
plt.gcf().autofmt_xdate()
plt.title("Valor investido na saúde por pessoa no mês dezembro de 2019 por UFs")
plt.show()
```

```python
ordenado_por_invest_pessoa_dez_2019['Invest/Pessoa'].loc[['41 Paraná','43 Rio Grande do Sul']].T.plot(kind='bar',figsize=(10,6)).grid()
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
plt.gcf().autofmt_xdate()
plt.title("Valor investido na saúde por pessoa no mês de dezembro de 2019")
plt.show()
```

**O Valor investido em saúde no Paraná é maior do que o do Rio Grande do Sul, mesmo com o número de população semelhante. No gráfico abaixo, mostra a difença entre o valor por pessoa e a população.**

```python
ax = ordenado_por_invest_pessoa_dez_2019['Invest/Pessoa'].loc[['41 Paraná','43 Rio Grande do Sul']].T.plot(kind='bar',color="lightgreen",figsize=(10,6))
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.set_ylabel("Valor por Pessoa")
ax2 = ax.twinx()
ax2.plot(ax.get_xticks(),ordenado_por_invest_pessoa_dez_2019['População_estimada'].loc[['41 Paraná','43 Rio Grande do Sul']].T.values /1000000, linestyle='-', marker='o', linewidth=2.0)
ax2.set_ylim(11,11)
ax2.set_ylabel("População (Milhões)")
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.1f}"))
plt.gcf().autofmt_xdate()
plt.title("Valor investido na saúde por pessoa em dezembro de 2019 e camparação entre as populações das 2 UFs")
plt.show()
```
